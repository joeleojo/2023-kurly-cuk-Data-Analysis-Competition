```python
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Market_kurly.csv')
df
```

```python
df['content'] = df['content'].str.replace('[^가-힣 ]', '', regex = True)
df
```

```python
# txt 파일에서 stopword를 읽어오기
with open('/content/stopwords.txt', 'r', encoding='UTF-8') as file:
    stopword_data = file.read()


# 쉼표로 구분된 데이터를 리스트로 변환
stopword_list = stopword_data.split(',')
stopword_list = pd.DataFrame(stopword_list)
stopword_list.head()
```

```python
import konlpy
okt = konlpy.tag.Okt()

review_tokenized = df['content'].apply(lambda review: okt.nouns(review))
```

```python
def remove_stopword(tokens):
    review_removed_stopword = []
    for token in tokens:
        # 토큰의 글자 수가 2글자 이상인 경우
        if 1 < len(token):
            # 토큰이 불용어가 아닌 경우만 분석용 리뷰 데이터로 포함
            if token not in list(stopword_list[0]):
                review_removed_stopword.append(token)

    return review_removed_stopword
```

```python
df
```

```python
df['nouns_removed_stopword'] = review_tokenized.apply(remove_stopword).replace(",", "")
nouns_removed_stopword = df['nouns_removed_stopword']
```

```python
nouns = nouns_removed_stopword.explode()
nouns
```

```python
df_word = pd.DataFrame({'word': nouns})
df_word['count'] = df_word['word'].str.len()
df_word = df_word.query('count >= 2')
df_word
```

```python
df_word = df_word.groupby('word', as_index = False).count().sort_values('count', ascending = False)
df_word
```

```python
dic_word = df_word.set_index('word').to_dict()['count']
dic_word
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 폰트 캐시 제거
plt.rcParams['font.family'] = 'Malgun Gothic'

# WordCloud 생성
wc = WordCloud(random_state=123, width=400, height=400, background_color='white')
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 이미지 플로팅
plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
plt.savefig('컬리 클라우드')
plt.show()
```
